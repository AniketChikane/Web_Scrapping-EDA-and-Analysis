```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.themoviedb.org/movie?page='
url2='https://www.themoviedb.org/movie'
```

```python
page_url=[]
for no in range (1,501):
    page_url.append(url+str(no))
```

```python
page_link=[]
for link in page_url[:51]:
    session = requests.Session()
    session.max_redirects = 60
    response=session.get(link,headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.67 Safari/537.36'}).text
    html_data=BeautifulSoup(response,'lxml')
    data_find=html_data.find_all('div',class_='wrapper')
    for i in data_find:
        serial=i.a['href'].replace('/movie','')
        page_link.append(url2+str(serial))
```

```python
page_link
j=1
Movie_Data=[]

for n in range(len(page_link)):
    session = requests.Session()
    session.max_redirects = 60
    resp=session.get(page_link[n],headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.67 Safari/537.36'}).text
    sup=BeautifulSoup(resp,'lxml')
    
    mname=sup.find('h2').text.strip()
    mname=mname.split('\n')[0]
    #print(mname)
    rating=str(sup.find('span',class_='icon'))
    rating=rating.split("icon-r")[1][:2]
    #print(rating)
    genre=sup.find('span',class_='genres').text
    genre=genre.strip()
    #print(genre)
    relase=sup.find('span',class_='release').text
    relase=relase.strip()
    #print(relase)
    run=sup.find('span',class_='runtime')
    run=str(run).replace('<span class="runtime">','')
    run=run.replace('</span>','')
    run=run.strip()
    #print(run)
    director=sup.find_all('li',class_='profile')
    lst=[]
    dict_name=[]
    for char in director:
        name=char.p.text
        char=char.text.strip().replace('\n',' ')
        lst=char.split(' ')
        for dct in lst:
            if (dct=='Director,' or dct=='Director'):
                dict_name.append(name)
    #print(dict_name)
    Movie_data={
        'Sr.No':j,
        'name':mname,
        'rating':rating,
        'genre':genre,
        'release date':relase,
        'runtime':run,
        'director':','.join(dict_name),
        'url':page_link[n]
    }
    #print(Movie_data)
    Movie_Data.append(Movie_data)
    #print(Movie_Data)
    j+=1
```

```python
df=pd.DataFrame(Movie_Data)
```

```python
df.to_csv('Movies data')
```

```python
def date(x):
    x=x[:11]
    return(x)
```

```python
df['release date']=pd.to_datetime(df['release date'].apply(date))
```

```python
def run(i):
    i=i.replace('m','')
    i=i.split()
    if len(i)==1:
        if (i[0][-1]=='h'):
            return(int(i[0][0])*60)
        elif(i[0]=='None'):
            i[0]=0
            return(i[0])
        else:
            return(i[0][0:])
    else:
        return((int(i[0][0]))*60 + int(i[1]))
```

```python
df['runtime']=df['runtime'].apply(run).astype(int)
```
